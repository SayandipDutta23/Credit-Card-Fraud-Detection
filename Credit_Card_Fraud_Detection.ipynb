# Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent transactions

Number of Fraudulaent Transactions ---> 103

```python
legit_sample = legit.sample(n = 103) #random sampling
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0) # '0' means row-wise , '1' means coloumn-wise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
# compare the values for both transactions
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2) # 0.2 means 20% of the data
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

❤ Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ',training_data_accuracy)
```

```python
# accuracy on the testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Testing data : ',test_data_accuracy)
```
